### Identifying join keys

If we are lucky, the three DataFrames will have columns with the same name that also contain the same data. If we are less lucky, we need to identify a (differently-named) column in each DataFrame that contains the same information.

The report number or ACRS Report Number (acrs_report_type) assigned to the incident is a common key identifier.

Now that we know the fields with the common incidents ID attributes in each DataFrame, we are almost ready to join our data. However, since there are different types of joins, we also need to decide which type of join makes sense for our analysis.

# Left joins
What if we want to add information from Drivers data and Non-Motorists data to Incidents data without losing any of the information from Incidents? In this case, we use a different type of join called a “left outer join”, or a “left join”.
a left join will return all of the rows from the left DataFrame, even those rows whose join key(s) do not have values in the right DataFrame. Rows in the left DataFrame that are missing values for the join key(s) in the right DataFrame will simply have null (i.e., NaN or None) values for those columns in the resulting joined DataFrame.

Note: a left join will still discard rows from the right DataFrame that do not have values for the join key(s) in the left DataFrame.

```python
#A left join is performed in pandas by calling the merge function, but using the how='left' argument:
#We will fist left join Drivers data to the Incidents data, then left join the Non-Motorists data to the sesult. 

df1 = pd.DataFrame(requests.get('https://data.montgomerycountymd.gov/resource/bhju-22kf.json').json())
df2 = pd.DataFrame(requests.get('https://data.montgomerycountymd.gov/resource/mmzv-x632.json').json())
merged_left1 = pd.merge(left=df1, right=df2, how='left', left_on='report_number', right_on='report_number')

merged_left1.shape
merged_left1

```

```python
df3 = pd.DataFrame(requests.get('https://data.montgomerycountymd.gov/resource/n7fk-dce5.json').json())
merged_left2 = pd.merge(left=merged_left1, right=df3, how='left', left_on='report_number', right_on='report_number')

merged_left2.shape
merged_left2
```

# Inner joins
The most common type of join is called an inner join. An inner join combines two DataFrames based on a join key and returns a new DataFrame that contains only those rows that have matching values in both of the original DataFrames.

Inner joins yield a DataFrame that contains only rows where the value being joined exists in BOTH datasets.

The pandas function for performing joins is called merge and an Inner join is the default option:



```python
# In this case `acrs_report_type` is the only column name in  the three dataframes, so if we skipped `left_on`
# And `right_on` arguments we would still get the same result
df1 = pd.DataFrame(requests.get('https://data.montgomerycountymd.gov/resource/bhju-22kf.json').json())
df2 = pd.DataFrame(requests.get('https://data.montgomerycountymd.gov/resource/mmzv-x632.json').json())
merged_inner1 = pd.merge(left=df1, right=df2, left_on='acrs_report_type', right_on='acrs_report_type')

# What's the size of the output data?
merged_inner1.shape
merged_inner1
```

```python
df3 = pd.DataFrame(requests.get('https://data.montgomerycountymd.gov/resource/n7fk-dce5.json').json())
merged_inner2 = pd.merge(left=merged_inner1, right=df3, left_on='report_number', right_on='report_number')

# What's the size of the output data?
merged_inner2.shape
merged_inner2
```
